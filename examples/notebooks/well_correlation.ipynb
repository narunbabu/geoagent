# Well Log Correlation Section

This notebook demonstrates how to create professional well log correlation
sections using GeoAgent's section plotter.

A correlation section shows multiple wells side-by-side, flattened on a
stratigraphic datum, with formation tops connected between wells.

````python
from geoagent import CoreDataManager, SectionPlotConfig, FormationTop

PROJECT_DIR = r'C:\Delete_ML_Projects\test2'  # Change to your project
dm = CoreDataManager(PROJECT_DIR)
print(dm)
````

## Define Section Configuration

Use `SectionPlotConfig` to define formation tops, datum surface, and display parameters.

````python
config = SectionPlotConfig(
    formation_tops={
        'K-VIII': FormationTop(color='green', label='K-VIII'),
        'K-VIII-A': FormationTop(color='blue', label='K-VIII-A'),
        'K-VIII-A-Lower': FormationTop(color='red', label='K-VIII-A-Lower'),
        'K-VIII-Coal': FormationTop(color='black', linewidth=2.0, label='K-VIII-Coal'),
    },
    datum_surface='K-VIII-A-Lower',
    interval_top_surface='K-VIII-A-Lower',
    interval_base_surface='K-VIII-Coal',
    window_above=35,
    window_below=25,
)
print(f'Datum: {config.datum_surface}')
print(f'Tops: {list(config.formation_tops.keys())}')
````

## Prepare Well Data

Use the log windower to extract and window log data for each well in the section.

````python
from geoagent.well.log_windower import prepare_section_data

# Define wells in the section (left to right)
section_wells = ['BK-14', 'BK-10', 'BK-2']  # Change to your wells

section_data = prepare_section_data(
    dm, section_wells, config
)

for wd in section_data:
    print(f"{wd['well_name']}: datum_md={wd.get('datum_md', 'N/A')}, "
          f"logs={'yes' if wd.get('logs') is not None else 'no'}")
````

## Plot the Section

````python
%matplotlib inline
from geoagent.plotting.section_plotter import plot_correlation_section

fig = plot_correlation_section(section_data, config)

# Save to file
fig.savefig('correlation_section.png', dpi=150, bbox_inches='tight')
print('Saved: correlation_section.png')
````

## Customization

Override any display parameter in `SectionPlotConfig`:

```python
config = SectionPlotConfig(
    gr_range=(0, 200),        # Wider GR range
    gr_sand_cutoff=60,        # Lower sand cutoff
    nphi_range=(0.60, -0.15), # Wide neutron range
    window_above=50,          # More depth above datum
    window_below=30,          # More depth below datum
)
```

See `geoagent.plotting.config` for all available parameters.
